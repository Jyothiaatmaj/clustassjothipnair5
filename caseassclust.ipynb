```python
x=data1.iloc[:,0:].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,13):
    kmeans =KMeans(n_clusters=i,init = 'k-means++',random_state=42)
    kmeans.fit(data1)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(range(1,13),wcss)
plt.title("the elbow method")
plt.xlabel('np of cluster')
plt.ylabel("WCSS")
plt.show()
```

```python
kmean = KMeans(n_clusters=3,init='k-means++',random_state=42)
```

```python
y_kmean = kmean.fit_predict(x)
```

```python
data2= pd.read_csv("/content/winequality-red.csv")
data2
```

```python
x2 = data2.iloc[:,:].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc2 = StandardScaler()
```

```python
x2 = sc2.fit_transform(x2)
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(10,6))
dendogram = sch.dendrogram(sch.linkage(x2,method='ward'))
ax = plt.gca()
ax.axes.xaxis.set_ticks([])
plt.title('dendogram')
plt.xlabel('Data points/observations')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
```

```python
y_hc=hc.fit_predict(x2)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
```

```python
data3= pd.read_csv("/content/winequality-red.csv")
data3
```

```python
wine_data = load_wine()
X = wine_data.data
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X_scaled)

```

```python
labels = dbscan.labels_
```

```python
print("Cluster labels:\n", labels)
```
